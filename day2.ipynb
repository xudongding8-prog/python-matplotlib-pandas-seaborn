# matplotlib.pyplot.scatter() 方法

## 功能说明
通过 `matplotlib.pyplot.scatter()` 方法绘制散点图和气泡图。

## 语法
```python
scatter(x, y, s=None, c=None, marker=None, alpha=None, ......)
```

## 参数说明

| 参数名称 | 说明 |
| :------- | :--- |
| **x, y** | 接收 array，表示 x 轴和 y 轴对应的数据 |
| **s** | 接收数值或一维的 array，指定点的大小。若传入一维 array 则表示每个点的大小，可绘制气泡图时使用，默认为 None |
| **c** | 接收数值或一维的 array，指定点的颜色。若传入一维 array 则表示每个点的颜色，默认为 None |
| **marker** | 接收特定 string，表示绘制点的类型，默认为 None |
| **alpha** | 接收 0-1 的小数，表示点的透明度，默认为 None |

## 使用示例

### 基本散点图
```python
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y)
plt.title('基本散点图')
plt.show()
```

### 自定义颜色和大小
```python
# 设置颜色和大小
plt.scatter(x, y, c='red', s=50, alpha=0.6)
plt.title('自定义颜色和大小')
plt.show()
```

### 气泡图
```python
# 使用不同大小表示第三个维度
sizes = np.random.randint(20, 200, 100)  # 随机大小
colors = np.random.randn(100)  # 随机颜色

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.colorbar()  # 显示颜色条
plt.title('气泡图')
plt.show()
```

### 不同标记样式
```python
markers = ['o', 's', '^', 'D', '*']
colors = ['red', 'blue', 'green', 'orange', 'purple']

for i, (marker, color) in enumerate(zip(markers, colors)):
    plt.scatter(x[i*20:(i+1)*20], y[i*20:(i+1)*20], 
                marker=marker, c=color, label=f'Group {i+1}')

plt.legend()
plt.title('不同标记样式的散点图')
plt.show()
```

## 常用参数组合

| 用途 | 参数设置 |
| :--- | :------- |
| 基本散点图 | `scatter(x, y)` |
| 彩色散点图 | `scatter(x, y, c=colors)` |
| 气泡图 | `scatter(x, y, s=sizes, c=colors)` |
| 透明散点图 | `scatter(x, y, alpha=0.5)` |
| 自定义标记 | `scatter(x, y, marker='^')` |

````python
# 详细注释版本 - 数据预处理步骤详解

import pandas as pd

# 1. 读取Excel文件
# pd.read_excel(): 读取Excel文件
# 参数说明：
#   - '可视化图表.xlsx': 文件路径
#   - sheet_name=3: 读取第4个工作表（索引从0开始，所以3表示第4个）
data3 = pd.read_excel('可视化图表.xlsx', sheet_name=3)
print("原始数据：")
print(data3)

# 2. 转置数据
# .T: 转置操作，将行和列互换
# 作用：如果原始数据是横向排列的，转置后变成纵向排列
data3 = data3.T
print("\n转置后的数据：")
print(data3)

# 3. 删除第一行
# [1:]: 切片操作，从索引1开始到结尾
# 作用：通常第一行是标题行，删除后保留数据行
data3 = data3[1:]
print("\n删除第一行后的数据：")
print(data3)

# 4. 重置索引
# reset_index(): 重置DataFrame的索引
# 参数说明：
#   - drop=False: 保留原来的索引作为新的一列，而不是删除
# 作用：将原来的索引变成普通列，并创建新的数字索引
data3 = data3.reset_index(drop=False)
print("\n重置索引后的数据：")
print(data3)

# 5. 重命名列
# rename(): 重命名DataFrame的列名
# 参数说明：
#   - columns={0:'用水量'}: 将列名为0的列重命名为'用水量'
#   - inplace=True: 直接在原DataFrame上修改，不创建新的副本
data3.rename(columns={0:'用水量'}, inplace=True)

# 6. 再次重命名列
# 将'index'列重命名为'月份'
# 这个'index'列是之前reset_index()操作产生的
data3.rename(columns={'index':'月份'}, inplace=True)
print("\n最终处理后的数据：")
print(data3)
````

# pandas 数据处理方法详解

## 1. pd.read_excel()
**功能**: 读取Excel文件
**语法**: `pd.read_excel(filepath, sheet_name=None, ...)`
**参数**:
- `filepath`: 文件路径
- `sheet_name`: 工作表名称或索引（0表示第一个工作表）
**返回值**: DataFrame对象

## 2. .T (转置)
**功能**: 转置DataFrame，行变列，列变行
**语法**: `df.T`
**作用**: 改变数据的排列方向
**示例**: 如果原始数据是横向的，转置后变成纵向

## 3. 切片操作 [1:]
**功能**: 选择DataFrame的行
**语法**: `df[start:end]`
**参数**:
- `[1:]`: 从索引1开始到结尾
- `[:5]`: 从开始到索引4
- `[2:8]`: 从索引2到索引7
**作用**: 删除不需要的行（通常是标题行）

## 4. reset_index()
**功能**: 重置DataFrame的索引
**语法**: `df.reset_index(drop=False, inplace=False)`
**参数**:
- `drop=False`: 保留原索引作为新列
- `drop=True`: 删除原索引
- `inplace=False`: 返回新DataFrame
- `inplace=True`: 直接修改原DataFrame
**作用**: 将索引变成普通列，创建新的数字索引

## 5. rename()
**功能**: 重命名DataFrame的列名或索引
**语法**: `df.rename(columns={}, index={}, inplace=False)`
**参数**:
- `columns={old_name: new_name}`: 重命名列
- `index={old_name: new_name}`: 重命名索引
- `inplace=True`: 直接修改原DataFrame
**作用**: 让列名更有意义，便于后续处理

## 数据预处理流程总结
1. **读取数据** → 获取原始数据
2. **转置数据** → 调整数据方向
3. **删除标题行** → 清理不需要的行
4. **重置索引** → 规范化索引
5. **重命名列** → 使用有意义的列名

### hist()方法
通过`matplotlib.pyplot.hist()`方法绘制直方图，其语法如下：
```python
hist(x, bins=None, density=None, histtype='bar', log=False, stacked=False, normed=None, ......)
```

### 参数说明

| 参数名称   | 说明                                                                 |
| ---------- | -------------------------------------------------------------------- |
| x          | 数据集，最终的直方图将对数据集进行统计                               |
| bins       | 统计的区间分布                                                       |
| density    | bool，默认为False，显示的是频数统计结果，为True则显示频率统计结果，和`normed`效果一致 |
| histtype   | 可选('bar'、'barstacked'、'step'、'stepfilled')之一，默认为'bar'，'step'使用的是梯状，'stepfilled'则会对梯状内部进行填充，效果与'bar'类似 |
| log        | bool，默认False，即坐标轴是否选择指数刻度                             |
| stacked    | bool，默认False，是否为堆积状图                                       |
| normed     | bool，默认为False，显示的是频数统计结果，为True则显示频率统计结果，和`density`效果一致 |

````python
data6=pd.read_excel('可视化图表.xlsx',sheet_name=6)
print(data6)
````

````python
# 创建图形窗口
plt.figure()

# 创建子图形
# 第一个子图形，用于绘制直方图
plt.subplot(2, 1, 1)
plt.hist(data6['temp'], edgecolor='k')
# 第二个子图形，用于绘制密度直方图（频率）
# 将每个bin的频率除以bin的宽度，以得出每个bin的密度值
plt.subplot(2, 1, 2)
plt.hist(data6['temp'], edgecolor='k', density=True)

# 显示图形
plt.show()
````

### pie()方法
通过`matplotlib.pyplot.pie()`方法绘制饼图，其语法如下：
```python
pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, textprops=None,……)
```

### 参数说明

| 参数名称    | 说明                                                                                                       |
| ----------- | ---------------------------------------------------------------------------------------------------------- |
| x           | 接收array，表示用于绘制饼图的数据                                                                          |
| explode     | 接收array，表示指定项离饼图圆心为n个半径，默认为None                                                       |
| labels      | 接收array，指定每一项的名称，默认为None                                                                    |
| color       | 接收特定string或包含颜色字符串的array，表示饼图的颜色，默认为None                                           |
| autopct     | 接收特定string，指定数值的显示方式，默认为None                                                              |
| pctdistance | 接收float，指定每一项的名称和距离饼图多少个半径，默认为0.6                                                  |
| shadow      | Bool，在饼图下是否绘制阴影，默认为False                                                                    |
| labeldistance| float或None，可选，默认值为1.1。绘制饼图标签的径向距离。如果设置为“None”，则不绘制标签，而是存储在legend（）中使用 |
| startangle  | float，可选，默认值为None，如果不是None，则从X轴逆时针方向旋转饼图的开头。                                 |
| radius      | 接收float，表示饼图的半径，默认为1                                                                         |
| textprops   | 接收dict，指定文本对象的参数，默认None                                                                      |

````python
import pandas as pd
import matplotlib.pyplot as plt

# 读取Excel文件中的指定工作表数据
data7 = pd.read_excel('可视化图表.xlsx', sheet_name=7)
# 打印数据
print(data7)

# 创建图形窗口
plt.figure()

# 绘制饼图
# 使用data7['金额(万元)']作为数据源
# 使用data7['区域']作为饼图的标签
# 使用autopct参数设置百分比的格式：automatic percentage。%.1f%%：保留一位小数并加入%号
# 使用textprops参数设置标签的颜色：text properties
# 使用explode参数设置饼图的突出效果
# 使用shadow参数设置饼图的阴影效果
# 使用startangle参数设置饼图的起始角度
plt.pie(data7['金额(万元)'], labels=list(data7['区域']), autopct='%.1f%%', textprops={'color': 'r'}, explode=[0.1, 0, 0, 0], shadow=True, startangle=90)

# 添加图例
plt.legend(loc=1) 

# 显示图形
plt.show()
````

## 注！
# 1
`plt.legend(loc=1)` 是用于设置图例的位置的代码。

在 `matplotlib` 库中，`plt.legend()` 函数用于添加图例。`loc` 参数用于指定图例的位置，可以使用不同的值来设置图例的位置。其中，`loc=1` 表示将图例放置在图形的右上角。如果不写loc值则自动挑选合适位置

具体来说，`loc` 参数可以接受以下值：

- `0`：上
- `1`：右上
- `2`：右
- `3`：右下
- `4`：下
- `5`：左下
- `6`：左
- `7`：左上
# 2
在 `textprops` 参数中，可以填写其他属性来设置文本的属性。以下是一些常用的属性：

- `color`：用于设置文本的颜色。可以是颜色名称（例如 `'red'`、`'blue'` 等）或颜色代码（例如 `'#FF0000'`、`'rgb(255, 0, 0)'` 等）。
- `fontsize`：用于设置文本的字体大小。可以是数字（例如 `10`）或数字和单位组合（例如 `'10px'`、`'10pt'` 等）。
- `fontfamily`：用于设置文本的字体系列。可以是字体名称（例如 `'Arial'`、`'Helvetica'` 等）或字体系列名称（例如 `'sans-serif'`、`'serif'` 等）。
- `fontstyle`：用于设置文本的字体样式。可以是 `'normal'`、`'italic'`、`'oblique'`、`'bold'`、`'bolder'`、`'lighter'` 等。
- `fontweight`：用于设置文本的字体粗细。可以是数字（例如 `400`）或字体粗细名称（例如 `'normal'`、`'bold'` 等）。
- `fontvariant`：用于设置文本的字体变体。可以是 `'normal'`、`'small-caps'` 等。

这些属性可以根据需要进行组合使用。例如，`textprops={'color': 'red', 'fontsize': 12, 'fontfamily': 'Arial'}` 表示将文本的颜色设置为红色，字体大小设置为12点，字体系列设置为Arial。
# 3
在 `plt.pie()` 函数中，`explode` 参数用于设置饼图的扇形的突出程度。`explode` 参数是一个列表，包含了每个扇形的突出程度。

具体来说，`explode` 参数的值是一个列表，其中每个元素对应一个扇形的突出程度。例如，`explode=[0.1, 0, 0, 0]` 表示将第一个扇形的突出程度设置为0.1，将其他三个扇形的突出程度设置为0。

在饼图中，扇形的突出程度可以用来突出某个扇形，或者使得某个扇形更加突出。例如，如果将某个扇形的突出程度设置为较大的值（例如0.2），那么该扇形将会突出其他扇形。


#### 1. 雷达图用途
雷达图常用于对多项指标的全面分析，能直观展示不同维度数据的对比或同一对象在多维度下的表现。

#### 2. 极坐标系基础
用 Matplotlib 绘制雷达图需依赖极坐标系，其核心定义如下：
- 在平面内取一个定点 O，该点称为**极点**。
- 从极点 O 引一条射线 Ox，这条射线称为**极轴**。
- 对于平面内任意一点 M：
  - 用 θ（读作“西塔”）表示从极轴 Ox 到线段 OM 的角度，θ 称为点 M 的**极角**。
  - 用 r 表示线段 OM 的长度，r 称为点 M 的**极径**。
  - 点 M 的位置可通过（θ，r）描述，这组数值称为点 M 的**极坐标**，由此建立的坐标系即为极坐标系。
- 通常情况下：
  - 极角 θ 的单位为 rad（弧度）或 °（度）。
  - 极径 r 的单位为 1（即默认长度单位）。



````python
import numpy as np
import matplotlib.pyplot as plt

# 创建 x 轴数据
# 创建一个从0到2π的等差数组，间隔为0.02
x = np.arange(0, 2 * np.pi, 0.02)

# 创建 y 轴数据
# 将 x 轴数据除以6，得到 y 轴数据
y = np.array(x) / 6

# 创建图形窗口
plt.figure()

# 创建子图1
# 创建一个子图，位置为121
ax1 = plt.subplot(121)

# 绘制 x 轴数据和 y 轴数据的曲线
# 使用 '--' 线型绘制曲线，线宽为2
ax1.plot(x, y, '--', lw=2)
# lw 等于 linewidth

# 创建子图2
# 创建一个子图，位置为122，投影类型为极坐标
ax2 = plt.subplot(122, projection='polar')

# 绘制极坐标下的 x 轴数据和 y 轴数据的曲线
# 使用 '--' 线型绘制曲线，线宽为2
ax2.grid(linestyle='--')  # 在极坐标图中添加网格线
ax2.plot(x, y, '--', lw=2)

# 显示图形
plt.show()
````


### 1. `x = np.arange(0, 2*np.pi, 0.02)`
- `np.arange` 是 NumPy 库生成等差数列的函数，格式为 `np.arange(起始值, 终止值, 步长)`（结果包含起始值，不包含终止值）。
- 这里：
  - 起始值是 `0`（从0开始）。
  - 终止值是 `2*np.pi`（`np.pi` 是圆周率π，约等于3.1416，所以 `2*np.pi` 约等于6.2832，即2π弧度，对应360度）。
  - 步长是 `0.02`（每次增加0.02）。
- 效果：生成一个从 `0` 到接近 `2π`（不包含 `2π`）、间隔为 `0.02` 的数组，包含约 `(2π)/0.02 ≈ 314` 个数据点。  
  这些数据可以理解为“角度”（单位：弧度），覆盖了完整的一圈（0到360度）。


### 2. `y = np.array(x) / 6`
- 先将 `x` 转换为 NumPy 数组（`np.array(x)`，这里 `x` 已经是数组，所以这一步可简化为 `y = x / 6`）。
- 然后将 `x` 中的每个元素都除以6，得到 `y` 数组。
- 效果：`y` 的值是 `x` 的1/6，例如：
  - 当 `x=0` 时，`y=0`；
  - 当 `x=2π`（约6.2832）时，`y≈6.2832/6≈1.047`。  
  这些 `y` 值可以理解为极坐标系中的“极径”（距离原点的长度）。


### 总结
这两行代码生成了一组“角度（x）”和“极径（y）”的数据，适合用于极坐标系绘图（比如雷达图、极坐标曲线）。例如，用这些数据绘图时，会得到一条从角度0开始、绕原点旋转一圈（360度）、极径逐渐增大到约1.047的曲线。

````python
# 极坐标的方向
#使用set_theta_direction()方法用于设置极坐标的方向，为1时为正方向，即逆时针的角度，为-1时为反方向，即顺时针。
plt.figure()
ax1 = plt.subplot(121,projection='polar')
ax1.plot(x,y,'--',lw=2)
ax2 = plt.subplot(122,projection='polar')
ax2.plot(x,y,'--',lw=2)
ax2.set_theta_direction(-1)
plt.show()
````

````python
# 角度网格线显示
# 使用set_thetagrids()方法设置极坐标角度网格线显示，
# 第一个参数为所要显示网格线的角度值列表，默认为0°、45°、90°、135°、180°、225°、270°、315°的网格线，
# 第二个参数为外侧的刻度标签
plt.figure()
ax1=plt.subplot(121,projection='polar')
ax1.plot(x,y,'--',lw=2)
ax2 = plt.subplot(122,projection='polar')
ax2.plot(x,y,'--',lw=2)
ax2.set_thetagrids(np.arange(0,360,60),['A','B','C','D','E','F'])
plt.show()
````

````python
# 极径网格线显示
# 使用set_rgrids ()方法设置极径网格线，参数值为所要显示网格线的极径值列表，最小值不能小于等于0
plt.figure()
ax1=plt.subplot(121,projection='polar')
ax1.plot(x,y,'--',lw=2)
ax2=plt.subplot(122,projection='polar')
ax2.plot(x,y,'--',lw=2)
ax2.set_rgrids(np.arange(0,1,0.3))
plt.show()
````

````python
# 极径标签位置
# 使用set_rlabel_position()方法设置极径标签显示位置，参数为标签所要显示在的角度
plt.figure()
ax1=plt.subplot(121,projection='polar')
ax1.plot(x,y,'--',lw=2)
ax2=plt.subplot(122,projection='polar')
ax2.set_rlabel_position(90)
ax2.plot(x,y,'--',lw=2)
plt.show()
````

````python
# 极径范围、最大值、最小值
# 使用set_rlim()方法设置显示的极径范围，参数为最大值和最小值
# 使用set_rmax()方法设置显示的极径最大值，该方法要在绘制完图像后使用才有效
# 使用set_rmin()方法设置显示的极径最小值，该方法要在绘制完图像后使用才有效
plt.figure()
ax1=plt.subplot(221,projection='polar')
ax1.plot(x,y,'--',lw=2)
ax2=plt.subplot(222,projection='polar')
ax2.set_rlim(0.6,1.2)
ax2.plot(x,y,'--',lw=2)
ax3=plt.subplot(223,projection='polar')
ax3.plot(x,y,'--',lw=2)
ax3.set_rmax(1.5)
ax4=plt.subplot(224,projection='polar')
ax4.plot(x,y,'--',lw=2)
ax4.set_rmin(0.5)
plt.show()
````

````python
# 示例，读取可视化图表.xlsx中的第12个表，绘制雷达图
data11=pd.read_excel('可视化图表.xlsx',sheet_name=11)
print(data11.head())
angles=np.arange(0,2*np.pi,2*np.pi/len(data11))
#绘制极坐标系
plt.figure()
ax=plt.subplot(111,projection='polar')
ax.grid(ls='--')
ax.set_thetagrids(np.arange(0,360,360/len(data11['分类'])),data11['分类'])
ax.plot(angles,data11['员工A'],label='员工A')
ax.plot(angles,data11['员工B'],label='员工B')
ax.legend()
plt.show() 
# 可以看到，雷达图中存在没有闭合的问题，因此，需
# 要增加最后一个点为起点的坐标，用以闭合雷达图
````

````python
# 调整雷达图
angles1=np.concatenate((angles,[angles[0]]))
score_A=np.concatenate((data11['员工A'],[data11['员工A'][0]]))
score_B=np.concatenate((data11['员工B'],[data11['员工B'][0]]))
plt.figure()
ax=plt.subplot(111,projection='polar')
ax.grid(ls='--')
ax.set_thetagrids(np.arange(0,360,360/len(data11['分类'])),data11['分类'])
ax.plot(angles1,score_A,label='员工A')
ax.plot(angles1,score_B,label='员工B')
plt.title('员工的能力评价图')
#填充颜色
ax.fill(angles1,score_A,alpha=0.25)
ax.fill(angles1,score_B,alpha=0.25)
ax.legend()
plt.show()
````
