````python
# 创建第一个图形

import matplotlib.pyplot as plt
import numpy as np # 提供数组数据结构，支持向量和矩阵运算，matplotlib 通常需要 numpy 数组作为数据输入
# 定义x轴数据 
# np.linspace(start, stop, num)
# 从1到50，生成100个数据点
x = np.linspace(1,50,100)
# 定义y轴数据
y = 3*x+1
plt.figure() # 创建图形
# 绘制‘线图’
plt.plot(x,y)
# 显示图形
plt.show()
````

`plt.figure()` 函数的各个参数（皆为可选）：

# `plt.figure()` 参数详解

## 1. **`num`** - 窗口ID
如果不提供就会自增
```python
plt.figure(num=1)  # 创建ID为1的窗口
plt.figure(num=2)  # 创建ID为2的窗口
plt.figure()       # 自动分配ID（通常是递增的）
```

## 2. **`figsize`** - 图像尺寸
整数元组，默认是无。提供整数元组则会以该元组为长宽。
```python
plt.figure(figsize=(8, 6))    # 宽8英寸，高6英寸
plt.figure(figsize=(12, 8))   # 宽12英寸，高8英寸
plt.figure()                  # 使用默认尺寸
```

## 3. **`dpi`** - 分辨率
```python
plt.figure(dpi=100)          # 100 DPI（较低分辨率）
plt.figure(dpi=300)          # 300 DPI（高分辨率，适合打印）
plt.figure()                 # 使用默认DPI
```

## 4. **`facecolor`** - 背景颜色
```python
plt.figure(facecolor='white')     # 白色背景
plt.figure(facecolor='lightblue') # 浅蓝色背景
plt.figure(facecolor='#FF0000')   # 红色背景（十六进制）
plt.figure()                      # 默认背景色
```

## 5. **`edgecolor`** - 边框颜色
```python
plt.figure(edgecolor='black')    # 黑色边框
plt.figure(edgecolor='red')      # 红色边框
plt.figure()                     # 默认边框色
```

# 常用组合
```python
# 高质量图像（适合论文）
plt.figure(figsize=(8, 6), dpi=300, facecolor='white')

# 演示用图像（较大尺寸）
plt.figure(figsize=(12, 8), dpi=100)

# 简单图像（使用默认设置）
plt.figure()
```

````python
# 实际使用示例

import matplotlib.pyplot as plt
import numpy as np

# 创建一个自定义的图形窗口
plt.figure(
    num=1,                    # 窗口ID
    figsize=(10, 6),          # 10英寸宽，6英寸高
    dpi=300,                  # 300 DPI
    facecolor='lightgray',    # 浅灰色背景
    edgecolor='blue'          # 蓝色边框
)

x = np.linspace(1, 50, 100)
y = 3*x + 1
plt.plot(x, y)
plt.show()
````

````python
# 自定义坐标轴刻度
x=np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(-2,4) # 设置x轴范围
plt.ylim(-5,15) # 设置y轴范围
plt.xlabel('x轴') # 设置x轴标签
plt.ylabel('y轴') # 设置y轴标签

new_ticks=np.linspace(-2,4,5)
plt.xticks(new_ticks)
plt.yticks([-5,0,5,10,15],
           ['very bad','bad','normal','good','very good'])
plt.show()
````

````python
# 设置线宽和样式
# linewidth 设置线宽
# linestyle 设置线型

# linewidth: 取值范围为0-10，默认1.5
# linestyle: 可取“-”，“--”，“-.”,":"四种，默认为“-”
x = np.linspace(1, 50, 100)
y = 3*x+1
plt.figure()
plt.plot(x,y,linewidth=1,linestyle='-')
plt.title('linewidth=1,linestyle=-')
plt.show()
plt.figure()
plt.plot(x,y,linewidth=2,linestyle='--')
plt.title('linewidth=2,linestyle=--')
plt.show()
plt.figure()
plt.plot(x,y,linewidth=3,linestyle='-.')
plt.title('linewidth=3,linestyle=-.')
plt.show()
plt.figure()
plt.plot(x,y,linewidth=4,linestyle=':')
plt.title('linewidth=4,linestyle=:')
plt.show()
````

````python
# legend(): 添加图例 内容在plot()中的label中设置
# 图例是图表中用于解释数据或符号的文本。
# 图例通常位于图表的角落，用于标识图表中的不同数据系列。
# 图例可以包含数据系列的名称、颜色、线型等信息。
# 图例有助于读者理解图表的含义和数据之间的关系。
# 图例通常使用图例框（legend box）来显示，图例框包含图例的文本和符号。
# 图例框通常位于图表的角落，可以通过调整图例框的位置来控制图例的显示位置。

plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置字体为SimHei黑体可以正常显示中文
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
x=np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
# loc设置图例位置
#loc参数选项如下：0: ‘best‘ 1: ‘upper right‘ 2: ‘upper left‘ 3: ‘lower left‘ 4: ‘lower right‘ 5: ‘right‘ 6: ‘center left‘
# 7: ‘center right‘ 8: ‘lower center‘ 9: ‘upper center‘ 10: ‘center'
plt.show()
````

## matplotlib marker 参数取值及意义

| marker取值 | 意义                 | marker取值 | 意义                 |
| :--------- | :------------------- | :--------- | :------------------- |
| o          | 圆圈                 | -          | 点                   |
| D          | 菱形                 | s          | 正方形               |
| h          | 六边形1              | *          | 星号                 |
| H          | 六边形2              | d          | 小菱形               |
| -          | 水平线               | v          | 一角朝下的三角形     |
| 8          | 八边形               | <          | 一角朝左的三角形     |
| p          | 五边形               | >          | 一角朝右的三角形     |
| ,          | 像素                 | ^          | 一角朝上的三角形     |
| +          | 加号                 | \|         | 竖线                 |
| None       | 无                   | X          | x                    |




````python
## 使用示例

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 11)
y = np.sin(x)

# 不同 marker 的效果
markers = ['o', 's', '^', 'D', '*', '+', 'X', 'h', 'p', '8']
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']


plt.figure(figsize=(12, 8))
# 使用enumerate()函数遍历markers和colors列表，同时获取索引和对应的值。
# 在循环中，使用enumerate()返回的索引i作为y轴的偏移量，使得每个marker的y轴位置不同。
for i, (marker, color) in enumerate(zip(markers, colors)):
    plt.plot(x, y + i*0.5, marker=marker, color=color, label=f'marker="{marker}"')


plt.legend()
plt.title('不同 marker 样式展示')
plt.show()
````

````python
# marker: 设置标记
x=np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
plt.plot(x,y1,marker='o',label = 'y1=2x+1')
plt.plot(x,y2,marker='*',label = 'y2=x^2')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
plt.show()
````

## matplotlib 颜色缩写及代表的颜色

| 颜色缩写 | 代表的颜色 |
| :------- | :--------- |
| b        | 蓝色       |
| g        | 绿色       |
| r        | 红色       |
| c        | 青色       |
| m        | 品红       |
| y        | 黄色       |
| k        | 黑色       |
| w        | 白色       |


````python
## 使用示例

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

# 使用颜色缩写
plt.plot(x, y, 'b-', label='蓝色实线')
plt.plot(x, y+1, 'r--', label='红色虚线')
plt.plot(x, y+2, 'g:', label='绿色点线')
plt.plot(x, y+3, 'k.', label='黑色点')

plt.legend(loc = "best")
plt.title('颜色缩写示例')
plt.show()

## 组合使用
# 颜色 + 线条样式 + 标记
plt.plot(x, y, 'ro-')  # 红色圆圈实线
plt.plot(x, y+1, 'bs--')  # 蓝色方形虚线
plt.plot(x, y+2, 'g^:')  # 绿色三角形点线
plt.show()
````

````python
x = np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
plt.plot(x,y1,label='y1=2x+1',color='c')
plt.plot(x,y2,label='y2=x^2',color='k')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
plt.show()
````

````python
x = np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
# 颜色 + 线条样式共同使用缩写
plt.plot(x,y1,'c-',label='y1=2x+1')
plt.plot(x,y2,'k--',label='y2=x^2')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
plt.show()
````

````python
# 多图分格显示
x = np.linspace(-5,5,50)
y1=2*x+1
y2=x**2
plt.figure()
plt.subplot(2,1,1) # 2行1列，第1个图
plt.plot(x,y1,'c--',label='y1=2x+1')
plt.title('y1=2x+1')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
plt.subplot(2,1,2) # 2行1列，第2个图
plt.plot(x,y2,'k-',label='y2=x^2')
plt.title('y2=x^2')
plt.legend(loc = "best") # 设置图例位置: 自动选择最佳位置
plt.tight_layout() # 调整子图之间的间距
plt.show()
````

````python
x=np.linspace(-5,5,50)
y=[2*x+1,x**2,3*x,x**2+1] # 4个函数
title_text=['y=2*x+1','y=x**2','y=3*x','y=x**2+1'] # 4个函数标题
plt.figure()
for i in range(4):
    plt.subplot(2,2,i+1) # 总共2行2列，从1开始，第i+1个图
    plt.plot(x,y[i])
    plt.title(title_text[i])

plt.tight_layout() # 自动调整子图之间的间距，使它们不重叠
plt.show()
````

````python
# 保存图像 savefig(),用法：savefig(fname, dpi=None, facecolor='w', edgecolor='w', orientation='portrait', papertype=None, format=None, transparent=False, bbox_inches=None, pad_inches=0.1, frameon=None, metadata=None)
# fname: 文件名
# dpi: 分辨率
# facecolor: 背景颜色
# edgecolor: 边框颜色
# orientation: 方向
# papertype: 纸张类型
# format: 格式
x=np.linspace(1,50,100)
y=3*x+1
plt.figure()
plt.plot(x,y)
plt.savefig('折线图.jpg')
plt.savefig('折线图.svg')
plt.show()
````
