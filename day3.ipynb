### Series
#### 基本概念
Series是Pandas中一种**一维带标签的数组型数据结构**，它可以存储任意数据类型（整数、浮点数、字符串、Python对象等），并且每个元素都有一个与之关联的标签（即索引）。可以把它想象成一列带“名字”的表格数据，或者是一个键值对（字典）的有序集合，但更强调数组的运算特性。

#### 主要特点
- **索引与值**：由两部分组成，`index`（索引，类似字典的键，可自定义，默认是从0开始的整数序列）和`values`（存储的实际数据，是一个NumPy数组）。
- **一维性**：只有一个维度，用于表示单一维度的序列数据，比如时间序列、单一特征的观测值等。
- **向量化操作**：支持NumPy风格的向量化运算，例如对所有元素进行加减乘除、逻辑判断等操作，无需循环。
- **标签对齐**：在进行运算时，会自动根据索引标签对齐数据，这是Pandas区别于纯NumPy数组的重要特性，能避免数据错位问题。

#### 创建示例


````python
import pandas as pd
# 从列表创建，默认索引
s1 = pd.Series([1, 3, 5, 7, 9])
print(s1)
# 自定义索引
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s2)
````

### DataFrame
#### 基本概念
DataFrame是Pandas中**二维的表格型数据结构**，可以看作是由多个Series（共享相同索引）按列组合而成的集合，或者是一个带行索引和列名的Excel表格/数据库表。它既有行索引（`index`），也有列索引（`columns`），能方便地处理结构化的表格数据。

#### 主要特点
- **二维结构**：由行和列组成，行表示观测样本，列表示不同的特征（或变量）。
- **多数据类型**：每一列可以是不同的数据类型（比如一列是字符串，一列是数值，一列是日期等）。
- **灵活的索引**：行索引和列索引都可自定义，方便按标签进行数据的选取、切片等操作。
- **数据对齐与缺失值处理**：同样支持基于标签的数据对齐，并且内置了丰富的缺失值（`NaN`）检测、填充、删除等处理方法。
- **丰富的操作**：支持按行、按列的切片、过滤、分组（`groupby`）、合并（`merge`）、连接（`concat`）等复杂的数据处理操作。

#### 创建示例

````python
import pandas as pd
# 从字典创建，键为列名，值为列数据
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['New York', 'London', 'Paris']
}
df = pd.DataFrame(data)
print(df)
# 也可自定义行索引
df_custom_index = pd.DataFrame(data, index=['a', 'b', 'c'])
print(df_custom_index)
````

````python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False
````

````python
data1=pd.read_excel('可视化图表.xlsx',sheet_name=1) 
# 这里的data1 是dataframe格式，可以直接使用pandas里的.plot进行绘图
data1
````

````python
# 创建一个 Pandas Series 对象，根据 '年份' 列的值将 '0' 分配给 '年份' 等于 2007 的行，否则将 '1' 分配给其他行
color_data = 'C' + pd.Series(np.where(data1['年份'] == 2007, '0', '1'))

# 打印 color_data 对象的值
print(color_data)

# 绘制一个散点图，显示 '电话次数' 和 '销量' 两列的数据
# 此处可以直接输入字符串，pandas会自动将字符串映射到数据集中的列名并取出数据
# 使用 color=color_data.values 参数将 color_data 对象的值用作颜色属性
# 将 '年份' 列等于 2007 的行的颜色设置为 'C0'，其他行的颜色设置为 'C1'
data1.plot('电话次数', '销量', kind='scatter', color=color_data.values, title='电话次数和销量散点图1')

# 绘制另一个散点图，显示 '电话次数' 和 '销量' 两列的数据
# 使用 c='电话次数' 参数将 '电话次数' 列的值用作颜色属性
# 将每个点的颜色根据 '电话次数' 列的值进行区分
data1.plot('电话次数', '销量', kind='scatter', c='电话次数', title='电话次数和销量散点图2')
````

# color的解释
在绘制散点图时，我们使用 color=color_data.values 参数将 color_data 对象的值用作颜色属性。Pandas 中的颜色表示采用的是一种简洁的格式在
 Matplotlib 中，`'C0'`、`'C1'` 等是**默认颜色循环（default color cycle）** 中的预定义颜色，属于一组常用的基础颜色，具体对应：

- **`'C0'`**：默认对应**蓝色（blue）**  
- **`'C1'`**：默认对应**橙色（orange）**  


### 补充说明：
1. **颜色循环规则**：  
   Matplotlib 预设了一组 10 种颜色（`'C0'` 到 `'C9'`），用于自动为多条曲线/图形分配颜色，避免手动指定。顺序为：  
   `C0`（蓝）→ `C1`（橙）→ `C2`（绿）→ `C3`（红）→ `C4`（紫）→ `C5`（棕）→ `C6`（粉）→ `C7`（灰）→ `C8`（浅蓝）→ `C9`（浅绿）。

2. **代码中的作用**：  
   代码通过 `np.where` 判断年份是否为 2007，将符合条件的点标记为 `'C0'`（蓝色），不符合的标记为 `'C1'`（橙色），从而在散点图中用两种颜色区分不同年份的数据。


如果需要自定义颜色，可以将 `'C0'`/`'C1'` 替换为具体颜色名称（如 `'blue'`、`'orange'`）或十六进制码（如 `'#0000FF'` 对应蓝色）。

### 但这里用的是pandas的绘图功能，为什么与matplotlib的颜色规则一致呢？？？

你说得对，这里用 **Pandas 的绘图功能**（`DataFrame.plot()`），但 Pandas 的绘图是**基于 Matplotlib 实现**的（可以理解为对 Matplotlib 的“封装”，让语法更简洁）。

所以，Pandas 绘图时使用的颜色规则（包括 `'C0'`、`'C1'` 这类命名），**完全继承自 Matplotlib 的默认颜色循环**～

简单来说：
Pandas 绘图是“上层接口”，底层渲染依赖 Matplotlib，因此颜色逻辑和 Matplotlib 是一致的——`'C0'` 对应蓝色，`'C1'` 对应橙色。

### 怎么没指定kind给我画了个line折线图？

在 Pandas 中使用 `Series.plot()` 或 `DataFrame.plot()` 方法时，如果不指定 `kind` 参数，**默认绘制的是折线图（line plot）**。


### 原理说明：
Pandas 的 `plot()` 方法中，`kind` 参数的默认值为 `'line'`，因此当不主动设置 `kind` 时，会自动按折线图的形式绘制数据。

折线图适合展示**数据随时间或有序类别变化的趋势**，例如：
- 时间序列数据（如每月销售额、每日气温）
- 连续变量的变化关系（如x递增时y的变化）


### 示例验证：

````python
import pandas as pd
import numpy as np

# 创建示例数据
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 3, 5, 6]
})

# 不指定kind，默认绘制折线图
data.plot(x='x', y='y')  # 等价于 data.plot(x='x', y='y', kind='line')
````

生成以 `x` 为横轴、`y` 为纵轴的折线图，与显式指定 `kind='line'` 的效果完全一致。

如果需要绘制其他类型的图（如散点图、柱状图），则必须通过 `kind` 参数指定（如 `kind='scatter'`、`kind='bar'`）。

在 Pandas 的 `DataFrame.plot()` 方法中，**默认会自动处理 `x` 和 `y` 的映射逻辑**，具体规则和你的示例场景有关：


### 核心逻辑：索引 + 列的自动映射
当你调用 `data4.plot(kind='bar')` 或 `data4.plot(kind='barh')` 时：
- **`x` 轴/垂直轴**：默认使用 DataFrame 的**索引（index）** 作为分类标签。在你的示例中，产品名称（产品A、产品B…）就是 DataFrame 的索引。
- **`y` 轴/水平轴**：默认使用 DataFrame 的**列（columns）** 作为数值来源。在你的示例中，“销售额(万元)”这一列就是数值来源。


### 更细节的解释
假设 `data4` 的结构是这样的（索引为产品，列是销售额）：
|        | 销售额(万元) |
| ------ | ------------ |
| 产品A  | 1000         |
| 产品B  | 800          |
| 产品C  | 500          |
| ...    | ...          |

- 当 `kind='bar'`（垂直柱状图）：
  - 索引（产品A、产品B…）会被自动映射到 **x 轴的分类标签**。
  - 列“销售额(万元)”会被自动映射到 **y 轴的数值高度**。

- 当 `kind='barh'`（水平柱状图）：
  - 索引（产品A、产品B…）会被自动映射到 **垂直轴的分类标签**。
  - 列“销售额(万元)”会被自动映射到 **水平轴的数值长度**。


### 什么时候需要手动指定 `x` 和 `y`？
如果你的 DataFrame 有**多列数值**，且需要选择“哪一列作为 y 轴”，或者**索引不是你想要的 x 轴标签**，才需要手动指定 `x` 和 `y`。

例如：
```python
# 假设data4有两列：销售额、利润
data4.plot(x='产品名称列', y='销售额', kind='bar')  # 手动指定x和y
```


### 总结
在你当前的示例中，因为 DataFrame 的**索引正好是“产品名称”（x 轴需要的分类标签）**，且**只有一列数值（销售额）**，所以 Pandas 会自动完成 `x` 和 `y` 的映射，不需要手动指定～

Pandas 的 `DataFrame.plot()` 方法能这样灵活绘制不同类型柱状图，是因为它**对 Matplotlib 绘图功能进行了封装**，结合自身数据结构（行列组织形式），让基于 DataFrame 的绘图变得简洁高效，下面从原理和参数作用来解释：

### 1. 数据结构与绘图的天然适配
Pandas 的 `DataFrame` 是**二维表格型数据结构**，每一列可以看作一个“系列（Series）”，行索引可作为分类标识（比如示例中的“季度”）。这种结构天然适合柱状图的绘制逻辑：
- 行索引（如“一季度”“二季度”）对应柱状图的**分类轴（x 轴）**。
- 每一列数据（如“产品A”“产品B”的数值）对应**不同系列的柱子**。


### 2. `kind='bar'`：基础柱状图（排列柱状图）
当设置 `kind='bar'` 时，Pandas 会为 `DataFrame` 的**每一列**绘制一组柱子，同一行索引（分类）下的不同列柱子**并排显示**，这就是“排列柱状图”。

以示例为例，`data5` 行是“季度”，列是不同产品，调用 `data5.plot(kind='bar')` 时：
- x 轴是“季度”（行索引）。
- 每一季度下，为“产品A”“产品B”等列分别绘制并排的柱子，展示不同产品在该季度的数值对比。


### 3. `subplots=True`：每列绘制子图
参数 `subplots=True` 的作用是：**为 `DataFrame` 的每一列单独绘制一个子图**。默认为一列多行，指定排列方式需要使用layout(行数,列数)，依次排列第一行A B第二行C D。

在示例中，`data5.plot(kind='bar', subplots=True)` 会针对“产品A”“产品B”等每一列，各自生成一个只包含该列数据的柱状图子图，方便单独观察每一列（单个产品）在各季度的数值变化。


### 4. `stacked=True`：堆叠柱状图
参数 `stacked=True` 会让同一行索引（分类）下的不同列柱子**垂直堆叠**，形成“堆叠柱状图”。

这样做的好处是：既能看到每一列（产品）的单独数值，又能直观展示同一分类（季度）下所有列数值的**总和**，以及各列数值在总和中的占比关系。


### 5. 对 Matplotlib 的封装：简化调用
Pandas 的 `plot()` 方法本质是调用 **Matplotlib** 的绘图函数（比如 `matplotlib.pyplot.bar`）。但它不需要你手动处理“如何从 DataFrame 中提取 x 轴数据、各系列数据”这些细节，而是**自动根据 DataFrame 的行列结构映射到绘图参数**，大大简化了代码。

比如，手动用 Matplotlib 实现堆叠柱状图，需要自己循环提取每一列数据、设置堆叠参数等，而 Pandas 只需一个 `stacked=True` 就可完成。


### 总结
Pandas 能这样便捷绘制不同柱状图，是**数据结构（DataFrame 行列组织）与绘图逻辑的适配**，加上对 Matplotlib 的高效封装，让“一行代码实现复杂绘图”成为可能～ 若你想更深入，也可以看看这些代码底层调用 Matplotlib 时的参数传递逻辑（不过日常使用中，掌握 Pandas 层面的接口就足够啦）。

````python
# 直方图
# 读取可视化图表.xlsx中的第7个表，绘制直方图
data6=pd.read_excel('可视化图表.xlsx',sheet_name=6)
print(data6.head())
````

````python
# 绘制基础直方图：展示海狸一天内的体温分布
# kind='hist' 表示绘制直方图，edgecolor='k' 为柱形添加黑色边框（k代表黑色）
data6.plot(kind='hist', title='海狸一天内体温分布情况', edgecolor='k')

# 注意：旧版本中用于将直方图纵坐标转换为密度的参数 'normed' 已被弃用
# 新版本中应使用 'density=True' 替代，效果一致（柱形面积总和为1，便于为频率/组距）
# 以下为错误示例（已注释）：
# data6.plot(kind='hist', title='海狸一天内体温分布情况', edgecolor='k', normed=1)

# 使用新版本参数 'density=True' 绘制密度直方图
data6.plot(kind='hist', title='海狸一天内体温分布情况', edgecolor='k', density=True)

# 设置 bins=5：指定直方图的柱形数量为5（ bins 控制数据分组数量，值越小分组越粗）
data6.plot(kind='hist', title='海狸一天内体温分布情况', edgecolor='k', bins=5)
````

````python
# 饼图 
# 读取Excel文件中索引为7的工作表（即第8个表，索引从0开始），将数据存入data7
data7 = pd.read_excel('可视化图表.xlsx', sheet_name=7)
# 将DataFrame的索引设置为“区域”列，方便后续基于区域进行饼图绘制
data7 = data7.set_index('区域')
# 打印data7的内容，查看数据结构
print(data7)
# 绘制饼图，kind='pie'指定绘制饼图；subplots=True表示为每一列绘制一个子图（这里因为只有一列“金额(万元)”，所以绘制一个饼图）；
# autopct='%1.1f%%'用于在饼图的每个扇区上显示百分比，保留1位小数
data7.plot(kind='pie', subplots=True, autopct='%1.1f%%')
````

````python
# 面积图
# 读取可视化图表中第九个表，绘制面积图
data8=pd.read_excel('可视化图表.xlsx',sheet_name=8)
data8 = data8.set_index('年份')
print(data8.head())
````

````python
# 绘制默认的堆叠区域图（各区域会相互堆叠，展示数值的累积效果）
data8.plot(kind='area')  

# 绘制区域图，设置透明度为0.5（使图形更具通透感，便于观察重叠部分），且设置 stacked=False（取消堆叠效果，各区域相互独立、不重叠，可更清晰对比单个区域的数值变化）
data8.plot(kind='area', alpha=0.5, stacked=False)  
````
