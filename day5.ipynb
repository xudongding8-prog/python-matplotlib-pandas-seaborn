### seaborn.heatmap()

热力图通常用来查看数据表里多个特征两两的相似度，类似于色彩矩阵，在Seaborn模块中，`heatmap()`函数可方便绘制热力图，语法如下：

```python
heatmap(data, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)
```

### 参数说明：
- **data**：矩阵数据集，可以是numpy的数组（array），如果是pandas的dataframe，则df的index/column信息会分别对应到heatmap的columns和rows；
- **vmax, vmin**：图例中最大值和最小值的显示值，没有该参数时默认不显示；
- **cmap**：从数字到色彩空间的映射，取值是matplotlib包里的colormap名称或颜色对象，或表示颜色的列表；
- **center**：数据表取值有差异时，设置热力图的色彩中心对齐值。通过设置center值，可以调整生成的图像颜色的整体深浅；
- **robust**：默认取值False；如果是False，且没设定vmin和vmax的值，热力图的颜色映射范围根据具有鲁棒性的分位数设定，而不是用极值设定；
- **annot**（annotate的缩写）：默认取值False；如果是True，在热力图每个方格写入数据；如果是矩阵，在热力图每个方格写入该矩阵对应位置数据；
- **fmt**：字符串格式代码，矩阵上标识数字的数据格式，比如保留小数点后几位数字；
- **annot_kws**：默认取值False；如果是True，设置热力图矩阵上数字的大小颜色字体；
- **square**：设置热力图矩阵小块形状，默认值是False；
- **xticklabels, yticklabels**：控制每行每列标签名的输出。默认值是auto，自动选择标签的标注间距，将标签名不重叠的部分(或全部)输出。如果是True，则以DataFrame的列名作为标签名；
- **mask**：控制某个矩阵块是否显示出来。默认值是None。如果是布尔型的DataFrame，则将DataFrame里True的位置用白色覆盖掉。
- **linewidths**：热力图中方格之间的线条宽度；
- **linecolor**：热力图中方格之间线条的颜色；
- **cbar**：是否显示颜色条（colorbar），默认True；
- **cbar_kws**：颜色条的额外参数，以字典形式传递；
- **cbar_ax**：指定颜色条绘制的轴（Axes）；
- **ax**：指定绘制热力图的轴（Axes）。

````python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.rcParams['font.sans-serif']=['SimHei'] #显示中文
plt.rcParams['axes.unicode_minus']=False #让负号正常显示
# 生成一个10行15列的随机整数数组，数值范围在1到99之间
x = np.random.randint(1, 100, (10, 15))
# 打印生成的数组（查看原始数据）
print(x)
# 使用seaborn绘制热力图，将数组x的数值分布用颜色表示
ax = sns.heatmap(x)
````

````python
# seaborn.heatmap()函数示例：自定义标注格式
# 设置随机种子，确保每次运行生成的随机数相同（结果可复现）
np.random.seed(100)
# 生成一个10行10列的随机浮点数数组，数值范围在0到1之间
x = np.random.rand(10, 10)
# 打印生成的数组（查看原始数据）
print(x)
# 绘制热力图并自定义标注：
# annot=True：在热力图的每个方格中显示对应数值
# annot_kws={'size':8,'color':'w'}：设置标注文字的样式（大小为8，颜色为白色）
# fmt='.2f'：设置数值格式为保留2位小数
ax = sns.heatmap(x, annot=True, annot_kws={'size':8, 'color':'w'}, fmt='.2f')
````

### 6.2 词云

#### WordCloud()函数
`wordcloud.WordCloud()`函数语法如下：
```python
WordCloud(font_path=None, width=400, height=200, margin=2, ranks_only=None, prefer_horizontal=0.9, mask=None, scale=1, color_func=None, max_words=200, min_font_size=4, stopwords=None, random_state=None, background_color='black', max_font_size=None, font_step=1, mode='RGB', relative_scaling='auto', regexp=None, collocations=True, colormap=None, normalize_plurals=True, contour_width=0, contour_color='black', repeat=False, include_numbers=False, min_word_length=0)
```

常用参数说明如下：
- **font_path**：指定字体的路径，默认为`None`。
- **width**：指定词云图片的宽度，默认为`400`。
- **height**：指定词云图片的高度，默认为`200`。
- **mask**：指定词云的形状，默认为长方形，也可以通过图形指定。
- **min_font_size**：最小字号，默认为四号。
- **max_font_size**：最大字号，默认为`None`。
- **font_step**：指定词云中词语步进间隔，默认为`1`。
- **max_words**：指定词云显示的最大单词数量，默认为`200`。
- **stop_words**：指定词云中排除词列表，即不显示的单词列表，对于中文不适用。
- **background_color**：指定图片的背景颜色。

### wordcloud词云完整步骤
wordcloud词云一般分为以下几个步骤：
1. 分隔：以空格分隔单词
2. 统计：单词出现的次数并过滤（去停用词）
3. 字体：根据统计搭配相应的字号（以显示中文）
4. 背景图片：准备一张黑白的背景图片，让词云按照图片进行显示
5. 绘制词云：通过`wordcloud.WordCloud()`、`generate()`、`to_file()`方法绘制词云

### jieba包的分词方法介绍

jieba 作为优秀的中文分词第三方库，提供了多种分词方法以适应不同场景需求，其中常用的有 `lcut()`、`cut()`、`cut_for_search()`，对应三种分词模式：

#### 1. `jieba.cut()` 方法
- **返回类型**：生成器（generator）。
- **作用**：对输入文本进行分词，逐个生成分词结果。
- **使用示例**：
  ```python
  import jieba
  text = "我喜欢用Python做数据分析"
  words = jieba.cut(text)
  # 需通过循环或转换为列表来获取分词结果
  print("/".join(words))  # 输出：我/喜欢/用/Python/做/数据/分析
  ```
- **特点**：节省内存，适合处理大规模文本，但不能直接通过索引访问分词结果，需要进一步操作。

#### 2. `jieba.lcut()` 方法
- **返回类型**：列表（list）。
- **作用**：对输入文本进行分词，并直接返回包含所有分词结果的列表。
- **使用示例**：
  ```python
  import jieba
  text = "我喜欢用Python做数据分析"
  words = jieba.lcut(text)
  print(words)  # 输出：['我', '喜欢', '用', 'Python', '做', '数据', '分析']
  ```
- **特点**：使用更便捷，可直接对返回的列表进行索引、切片等操作，适合大多数常规分词场景。

#### 3. `jieba.cut_for_search()` 方法
- **返回类型**：生成器（generator）。
- **作用**：适用于搜索引擎构建倒排索引的分词，会对文本进行更细粒度的拆分，产生更多更细的分词结果，以提高搜索的召回率。
- **使用示例**：
  ```python
  import jieba
  text = "我喜欢用Python做数据分析"
  search_words = jieba.cut_for_search(text)
  print("/".join(search_words))  
  # 输出可能更细粒度，比如：我/喜欢/用/Python/做/数据/分析（具体结果因jieba版本和词典可能略有差异，对于该示例可能和cut结果相近，复杂长文本差异更明显）
  ```
- **特点**：分词粒度更细，能挖掘出更多潜在的搜索关键词，专为搜索场景优化。

````python
# 导入需要的库：
# matplotlib.pyplot 用于图形显示
# WordCloud 用于生成词云
# jieba 用于中文分词
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba

# 打开文本文件：
# '成熟的面对敌人.txt' 是目标文本文件路径
# 'r' 表示只读模式打开
# encoding='gbk' 指定文件编码为gbk（解决中文解码问题）
f = open('成熟的面对敌人.txt', 'r', encoding='gbk')

# 读取文件内容，将整个文本读取为一个字符串
sentence = f.read()

# 关闭文件，释放系统资源
f.close()

# 数据清洗：去除文本中的空格（包括空格符）
sentence = sentence.replace(' ', '')

# 数据清洗：去除文本中的换行符（\n）和回车符（\r）
# 避免这些特殊字符干扰后续分词和词云生成
sentence = sentence.replace('\n', '').replace('\r', '')
````

````python
# 定义一个空列表stop_words，用于存储停用词
stop_words = []
# 以utf-8编码、只读模式打开名为“stopword.txt”的停用词文件
for line in open('stopword.txt', 'r', encoding='utf-8'):
    # 去除每行停用词的首尾空白符（包括换行符），并添加到stop_words列表中
    stop_words.append(line.strip('\n'))

# 定义一个空字符串word_str，用于存储分词后且去除停用词的词语（用空格分隔）
word_str = ""
# 使用jieba的lcut方法对文本sentence进行分词，得到分词结果列表word_cut
word_cut = jieba.lcut(sentence)
# 遍历分词结果列表中的每个词语
for word in word_cut:
    # 如果词语不在停用词列表中
    if word not in stop_words:
        # 将词语添加到word_str中，并用空格分隔
        word_str += word + ' '
````

````python
# 绘制词云
# 创建WordCloud对象，配置词云生成的相关参数
wc = WordCloud(
    font_path='simhei.ttf',  # 指定中文字体路径，确保中文能正常显示
    background_color='white',  # 设置词云背景颜色为白色
    width=1000,  # 设置词云图片的宽度为1000像素
    height=600,  # 设置词云图片的高度为600像素
    mask=plt.imread('cloud.jpg')  # 读取“cloud.jpg”图片作为词云的形状掩码，使词云呈现该图片的形状
)
wc.generate(word_str)  # 根据处理好的文本字符串word_str生成词云
wc.to_file('wordcloud.png')  # 将生成的词云保存为“wordcloud.png”图片文件
````
